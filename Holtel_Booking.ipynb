```python
# eliminamos las columnas agent y company -- no son relevantes y tienen valores NaN
df.drop(columns=['agent','company'],axis=1,inplace=True)
# eliminamos valores resto de na
df.dropna(inplace=True)
```

```python
# creo una columna datetime con la fecha de arrival a partir de las columnas arrival

months = {
    'january': 1, 'february': 2, 'march': 3, 'april': 4,
    'may': 5, 'june': 6, 'july': 7, 'august': 8,
    'september': 9, 'october': 10, 'november': 11, 'december': 12
}

df['month']=df['arrival_date_month'].str.lower().map(months)
df['arrival_date_dt']=pd.to_datetime({
    'year':df['arrival_date_year'],
    'month':df['month'],
    'day':df['arrival_date_day_of_month']
})
```

```python
# vamos a cambia la columna de posiciÃ³n
date_dt=df.pop('arrival_date_dt')
df.drop(columns='month',index=1,inplace=True)
```

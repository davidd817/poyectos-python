```python
import random # para generar numeros aleatorios en las slots


MAX_LINES=3 # establecemos una constante invariables para el numero de lineasmaximo de la slot
MAX_BET=100 # apuesta máxima
MIN_BET=1 # apuesta mínima

ROWS=3 # numero de filas de la slot
COLS=3

# vamos a crear una slot con cuantro 'símbolos por columna': ABCD

symbol_count={  # vamos a crear una slot con un numero determinado de simbolos de cada tipo
    'A':2,
    'B':4,
    'C':6,
    'D':8
}

def get_slot_machine_spin(rows,cols,symbols):
    all_symbols=[] #creamos una lista que tenga todos los simbolos disponibles para posteriormente elegir de manera random
    for symbol, symbol_count in symbols.items:
        for _ in range(symbol_count): # ponemos varible anónima _ ; necesito cada simbolo el numero de veces fijado en symbol_count
            all_symbols.append(symbol)
    columns=[]
    for col in range(cols):
        colum=[]
        current_symbols=all_symbols[:]
        for row in range(rows):
            value=random.choice(all_symbols)
            current_symbols.remove(value)
            colum.append(value)

        columns.append(column)
    return columns


def print_slot_machine(columns):
    for row in range(len(columns[0])):
        for i, column in enumerate(columns):
            if i != len(columns)-1:
                print(column[row], end=' | ')
            else:
                print(column[row], end=' | ')
        print()


def deposit(): # collecting users deposit
    while True:
        amount=input('What would you like to deposit? $')
        if amount.isdigit():
            amount=int(amount)
            if amount > 0:
                break
            else:
                print('Amount must be greater than 0')
        else:
            print('Please enter a number')
    return amount


def get_number_of_lines(): # apuesta por línea
    while True:
        lines=input('Enter number of lines to bet on (1-' + str(MAX_LINES) + ') ?')
        if lines.isdigit():
            lines=int(lines)
            if 1 <= lines <= MAX_LINES:
                break
            else:
                print('Enter a valid number')
        else:
            print('Please enter a number')
    return lines


def get_bet():
    while True:
        amount=input('What would you like to bet on each line? $')
        if amount.isdigit():
            amount=int(amount)
            if MIN_BET <= amount <= MAX_BET:
                break
            else:
                print(f'Amount must be between {MIN_BET} and {MAX_BET}')
        else:
            print('Please enter a number')
    return amount
    

def main():
    balance=deposit()
    lines=get_number_of_lines() # debemos comporbar que la apuesta no se mayor al balance
    while True:
        bet=get_bet()
        total_bet=lines*bet
        if total_bet > balance:
            print(f'You do not have enough to bet that amount, your current balance is: {balance}')
        else:
            break
    print(f'You are betting {bet} on {lines} lines. Total bet is equal to {total_bet}')

    slots=get_slot_machine_spin(ROWS,COLS,symbol_count)
    print_slot_machine(slots)




main()
```
